```python
import torch # Deep Learning framework required for most Hugging Face (HF) models
from transformers import AutoModelForCausalLM, AutoTokenizer #AutoModel to load pretrained model. Selects right model based on name

torch.set_default_device("cuda") #Use GPU if available
# Using Microsoft phi-2
model = AutoModelForCausalLM.from_pretrained("microsoft/phi-2", torch_dtype="auto", device_map="auto", trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-2", trust_remote_code=True)
```

```python
# Google colab form fields
prompt = "Where do apples grow from" # @param {type:"string"}
max_length = 256 # @param {type: "integer"}
inputs = tokenizer(prompt, return_tensors="pt", return_attention_mask=False)
outputs = model.generate(**inputs, max_length=max_length)
text = tokenizer.batch_decode(outputs)[0]
print(text.split("<|endoftext|>")[0])
```
